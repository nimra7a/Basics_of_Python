# Handling Missing Values

```python
data_mv = pd.read_csv('Sales_MV.csv')
```

```python
data_mv
```

```python
data_mv['Sales_Before']
```

```python
data_mv['Sales_Before'].mean()
```

# Filling NAN with mean value & change in original dataset with inplace=True

```python
data_mv['Sales_Before'].fillna(data_mv['Sales_Before'].mean(), inplace = True)
```

```python
data_mv['Sales_Before']
```

```python
data_mv
```

# or reassign the fillna with mean to the original one

# Removing the all NAN values/rows

```python
data_mv.dropna()
```

```python
data_mv #but still showing in original data
```

```python
# to overcome this probelm:either use the inplace = True or reassign to the original data
```

```python
data_mv.dropna(inplace=True)
```

```python
data_mv
```

```python
data_mv.to_csv('CleanedData.csv', index = False)
```
